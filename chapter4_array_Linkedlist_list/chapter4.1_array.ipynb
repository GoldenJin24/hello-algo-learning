# 4.1 数组
- 定义：作为一种线性的数据结构，在连续的内存空间中储存内容
- 基本操作：初始化，访问，插入，删除，遍历，查找，扩容
- 应用：一对一的查找表例如ASCII码，深度学习中的张量，实现更高级的数据结构，随机访问很快
- 优点：访问快，缓存局部性：当访问数组元素时，计算机不仅会加载它，还会缓存其周围的其他数据，从而借助高速缓存来提升后续操作的执行速度
- 缺点：消耗内存，插入删除扩容麻烦

```python
#初始化
array1 = [0] * 5
print(array1)
nums1 = [1,2,3,4,7]
print(nums1)
```

```python
#随机访问
import random
# randint 是左闭右闭
def random_access(nums):
    random_index = random.randint(0,len(nums)-1)
    return nums[random_index]

print(random_access(nums1))
```

```python
#插入一个元素
def insert(nums,num,index):
    #range 是左闭右开
    for i in range(len(nums)-1,index,-1):
        nums[i] = nums[i-1]
    nums[index] = num
    return nums
print(nums1)
print(insert(nums1,6666,1))
```

```python
#删除一个元素
#要遍历最后一个则是len(),不遍历最后一个则是len()-1
def remove(nums,index):
    for i in range(index,len(nums)-1):
        nums[i] = nums[i+1]
    nums[len(nums)-1] = 0
    return(nums)
print(nums1)
print(remove(nums1,2))
```

```python
#遍历数组
print(nums1)
for i,j in enumerate(nums1):
    print(i,j)
```

```python
#查找数组
#只能找到第一个一样的，如果要全部都找的话，for循环不用中途return，list[]一直append()即可
def find(nums,num):
    for i,j in enumerate(nums):
        if j == num:
            return i
    return None
print(nums1)
print(find(nums1,1012512))
print(find(nums1,6666))
```

```python
#扩容数组
#感觉平时应该用的很少吧？多次扩容是很浪费时间的，大部分情况宁愿一开始就创建一个大数组。
def extend(nums,enlarge_length):
    new_nums = [0] * ( int(len(nums)) + int(enlarge_length) )
    for i,j in enumerate(nums):
        new_nums[i] = j
    return new_nums
print(nums1)
print(extend(nums1,10000))
```

# 总结
- 代码能力有待提升，函数内的变量名记得不要写错，记得写return
- 特别是for循环，range()是左闭右开，random.randint()是左闭右闭
